# create new image

```python
import json
file = open('config.json')
config_env = json.load(file)

PIPELINE_PROJECT_ID       = config_env["PIPELINE_PROJECT_ID"]
PIPELINE_REGION           = config_env["PIPELINE_REGION"]
BASE_CONTAINER_IMAGE_NAME = config_env["PIPELINE_METADATA"]["value"]
PIPELINE_REPO             = 'repo-'+BASE_CONTAINER_IMAGE_NAME

BASE_IMAGE                = '{}-docker.pkg.dev/{}/{}/{}:'.format(PIPELINE_REGION, PIPELINE_PROJECT_ID, PIPELINE_REPO, BASE_CONTAINER_IMAGE_NAME)+'latest'
```

```python
BASE_IMAGE
```

```python
# For artifact repositories
# Crea un repositorio nuevo de Artifact Registry llamado containers:
#!gcloud artifacts repositories create $PIPELINE_REPO --repository-format=docker --location=$PIPELINE_REGION
# Compila y publica la imagen de contenedor:
!gcloud builds submit -t $BASE_IMAGE .
```
